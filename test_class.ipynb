```python
from PyECLOUDplots import PyECLOUDParameterScan
paramscan = PyECLOUDParameterScan("C:\\Users\\smbso\\cernbox\\simulations\\most_relevant")
sim_configs = paramscan.read_yaml_to_dict("./yaml_repo/simulations.yaml")
manget_conf = paramscan.read_yaml_to_dict("./yaml_repo/magnet_configs.yaml")
sim_colors = paramscan.read_yaml_to_dict("./yaml_repo/sim_colors.yaml")
```

```python
def get_x(sim):
    return sim.t/sim.b_spac
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = paramscan.plot_simulation_attribs(get_x,"Nel_timep",curves="SEY",curve_vals=np.arange(1.0,1.6,0.1),common_params=sim_configs["Cu2O Conditioned"]|manget_conf["Dipoles"]|{"Intensity":1.6},xlabel="Bunch Passage",title="Number of electrons in chamber as a function of SEY for 1.6 Intensity",ylabel=r"$N_{\rm el}$",cmap_max_offset=0.05,lw=4,show_datapoints=False,title_fontsize=18,returnfig=True)
plt.ylim(bottom=5*10**6,top=5*10**9)
plt.semilogy()
plt.show()
```

```python
fig = paramscan.plot_simulation_attribs(get_x,"Nel_timep",curves="Intensity",curve_vals=np.arange(0.2,1.8,0.2),common_params=sim_configs["Cu2O Conditioned"]|manget_conf["Dipoles"]|{"SEY":1.35},xlabel="Bunch Passage",title="Number of electrons in chamber as a function of Intensity for 1.35 SEY",ylabel=r"$N_{\rm el}$",cmap_max_offset=0.05,show_datapoints=False,lw=4,title_fontsize=18,returnfig=True)
plt.ylim(bottom=10**5,top=5*10**9)
plt.semilogy()
plt.show()
```
